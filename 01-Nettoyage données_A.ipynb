```python
def ColonnesChoisies(listecol,nouveauDF, dfDeBase='livres'):
    nouveauDF=pd.DataFrame()
    for lacolonnequongarde in livres.columns:
        jeton=False
        for colonne in listecol:
            if ((lacolonnequongarde==colonne)) :
                jeton=True
        if jeton==True :
            nouveauDF[lacolonnequongarde]=livres[lacolonnequongarde]
    return nouveauDF
```

```python
LstColonneChoisies=['ISBN','Langue','Titre','Editeur','Date','Format','Collection','Titre de série','Auteur Nom','Auteur Prénom','Auteur Dates','EAN','Catégorie statistique 1','Catégorie statistique 2']
DOCS_genres=ColonnesChoisies(LstColonneChoisies,"livres_genres")
DOCS_genres['Type']=livres.iloc[:,33]
```

```python
# Création d'un Dataset plus léger
# Danx ce fichier, on a tous les documents et que les voulus
DOCS_genres.to_csv("CSV_TMP/TousLesDocs.csv",sep="\t")
```
